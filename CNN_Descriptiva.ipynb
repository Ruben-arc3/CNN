```python

# CNN + GPT-2 Descripción automática (Español)
# Clasificación de señales de tránsito
# Ruben Mora Martinez

import os
import xml.etree.ElementTree as ET
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# =============================================================
# CONFIGURACIÓN DEL DATASET
# =============================================================

# Asume que DATASET_PATH, IMG_DIR, ANN_DIR existen y son accesibles
DATASET_PATH = r"../datasets"
IMG_DIR = os.path.join(DATASET_PATH, "images")
ANN_DIR = os.path.join(DATASET_PATH, "annotations")

CLASSES = ["trafficlight", "stop", "speedlimit", "crosswalk"]
IMG_SIZE = (96, 96) 

# =============================================================
# FUNCIÓN PARA CARGAR PASCAL VOC
# =============================================================

def load_voc_dataset(img_dir, ann_dir):
    images, labels, errores = [], [], []

    for xml_file in os.listdir(ann_dir):
        if not xml_file.endswith(".xml"):
            continue

        xml_path = os.path.join(ann_dir, xml_file)

        try:
            tree = ET.parse(xml_path)
            root = tree.getroot()

            filename = root.find("filename").text
            img_path = os.path.join(img_dir, filename)

            if not os.path.exists(img_path):
                errores.append(f"Imagen no encontrada: {filename}")
                continue

            img = cv2.imread(img_path)
            if img is None:
                errores.append(f"Error leyendo: {filename}")
                continue

            # Convertir a RGB y obtener dimensiones
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            h, w = img.shape[:2]

            for obj in root.findall("object"):
                cls = obj.find("name").text.lower().strip()

                if cls not in CLASSES:
                    continue

                box = obj.find("bndbox")
                xmin = int(float(box.find("xmin").text))
                ymin = int(float(box.find("ymin").text))
                xmax = int(float(box.find("xmax").text))
                ymax = int(float(box.find("ymax").text))

                # Asegurar que las coordenadas estén dentro de los límites de la imagen
                xmin, ymin = max(0, xmin), max(0, ymin)
                xmax, ymax = min(w, xmax), min(h, ymax)

                # Verificar dimensiones válidas del bounding box
                if xmax <= xmin or ymax <= ymin:
                    continue

                crop = img[ymin:ymax, xmin:xmax]
                if crop.size == 0:
                    continue

                # Redimensionar y guardar
                crop = cv2.resize(crop, IMG_SIZE)
                images.append(crop)
                labels.append(cls)

        except Exception as e:
            errores.append(f"[Error] Procesando {xml_file}: {e}")

    print(f"Imágenes cargadas: {len(images)}")

    if errores:
        print("\nEjemplos de errores:")
        print("\n".join(errores[:5]))

    return np.array(images), np.array(labels)


# =============================================================
# CARGAR DATASET
# =============================================================
X, y = load_voc_dataset(IMG_DIR, ANN_DIR)

# Mapeo y Normalización
label_to_idx = {cls: i for i, cls in enumerate(CLASSES)}
y_idx = np.array([label_to_idx[label] for label in y])

X = X / 255.0

# Split de datos
X_train, X_test, y_train, y_test = train_test_split(
    X, y_idx, test_size=0.2, random_state=42, stratify=y_idx
)

# =============================================================
# AUMENTO DE DATOS
# =============================================================
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)
datagen.fit(X_train)

# =============================================================
# MODELO CNN (ORIGINAL)
# =============================================================
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(*IMG_SIZE, 3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.4),

    layers.Dense(len(CLASSES), activation='softmax')
])

model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, verbose=1)

# =============================================================
# ENTRENAMIENTO
# =============================================================
# NOTA: Comenta esta sección si quieres correr solo la parte del LLM sin reentrenar.
print("\nIniciando entrenamiento...")
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=32),
    validation_data=(X_test, y_test),
    epochs=25,
    callbacks=[early_stop, reduce_lr]
)
print("Entrenamiento finalizado.")

# =============================================================
# EVALUACIÓN
# =============================================================
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"\nPrecisión en test: {test_acc:.2%}")
print(f"Pérdida en test: {test_loss:.4f}")

# =============================================================
# GRÁFICAS
# =============================================================
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del modelo')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida del modelo')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.legend()
plt.grid(True)

plt.show()

# =============================================================
# PREDICCIÓN CON UNA SOLA IMAGEN
# =============================================================
idx = random.randint(0, len(X_test) - 1)
img = X_test[idx]
img_batch = np.expand_dims(img, axis=0)

pred = model.predict(img_batch)
pred_class = np.argmax(pred)
pred_label = CLASSES[pred_class]

plt.imshow(img)
plt.title(f"Predicción del modelo: {pred_label}")
plt.axis('off')
plt.show()

print(f"Clase real: {CLASSES[y_test[idx]]}")
print(f"Clase predicha: {pred_label}")

# =============================================================
# GPT-2 EN ESPAÑOL PARA DESCRIPCIÓN AUTOMÁTICA (SECCIÓN MEJORADA)
# =============================================================

# Mapeo de la descripción base (Versión simple para no confundir al LLM)
descripcion_map = {
    "stop": "Señal reglamentaria de detención obligatoria (STOP).",
    "speedlimit": "Señal reglamentaria de límite de velocidad máxima.",
    "crosswalk": "Señal preventiva de cruce peatonal.",
    "trafficlight": "Señal reglamentaria para regular el tránsito con luces de colores."
}

descripcion_base = descripcion_map.get(pred_label, "Señal de tránsito.")

# Determinar el dispositivo
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Cargando GPT-2 en Español ({device})...")

# Modelo GPT-2 pequeño en español
MODEL_NAME = "datificate/gpt2-small-spanish"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model_gpt2 = AutoModelForCausalLM.from_pretrained(MODEL_NAME).to(device)


# =============================================================
# FUNCIÓN PARA GENERAR DESCRIPCIÓN CON GPT-2 (CON MÁS RESTRICCIONES Y RESPALDO)
# =============================================================

def generar_descripcion_gpt2(texto_base):
    # Prompt muy simple que pide 3 puntos clave y añade la instrucción de terminación.
    prompt = (
        f"Describa la siguiente señal de tránsito de manera técnica en un solo párrafo: '{texto_base}'. "
        "Incluya: 1) Clasificación (Reglamentaria/Preventiva), 2) Función principal, y 3) Acción requerida del conductor. "
        "El incumplimiento lleva a una multa y riesgo de accidente. RESUMEN COMPLETO: "
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    # Parámetros ajustados para MÁXIMA coherencia y mínima divagación.
    output = model_gpt2.generate(
        **inputs,
        max_length=150, 
        temperature=0.4, # Muy baja temperatura para forzar la respuesta más probable y lógica.
        top_k=50,
        top_p=0.85, # Reducimos la diversidad de palabras permitidas
        repetition_penalty=1.5,
        do_sample=True,
        num_return_sequences=1,
        # Agregamos un token de parada explícito (punto) para intentar terminar la frase a tiempo
        eos_token_id=tokenizer.encode('.')[0], 
        pad_token_id=tokenizer.eos_token_id 
    )

    respuesta = tokenizer.decode(output[0], skip_special_tokens=True)

    # Limpieza: Intentamos encontrar el marcador 'RESUMEN COMPLETO:' y quitar todo antes.
    if "RESUMEN COMPLETO:" in respuesta:
        respuesta = respuesta.split("RESUMEN COMPLETO:")[1].strip()
    
    # Limpieza de la frase base si persiste
    respuesta = respuesta.replace(texto_base, '').strip()

    # --- Mecanismo de Respaldo ---
    # Si la respuesta generada sigue siendo incoherente (ej: muy corta o sin puntos clave), 
    # se proporciona una descripción de respaldo manual y concisa.
    respaldo_manual = {
        "stop": "Esta es una **señal reglamentaria (STOP)** que exige la detención total del vehículo, siendo su función ceder el paso. El conductor debe detenerse completamente y solo avanzar cuando sea seguro. No respetarla implica riesgo de accidente y multa.",
        "speedlimit": "Esta es una **señal reglamentaria** que indica la **velocidad máxima** permitida en la vía. El conductor debe ajustar su velocidad para no exceder el límite. No acatarla es motivo de multa y representa un alto riesgo de accidente.",
        "crosswalk": "Esta es una **señal preventiva** que advierte la proximidad de un **cruce peatonal**. El conductor debe reducir la velocidad y estar listo para frenar. No ceder el paso al peatón puede causar un accidente y sanción.",
        "trafficlight": "Este es un **dispositivo reglamentario (semáforo)** cuya función es regular el flujo vehicular mediante luces. El conductor debe obedecer la luz encendida (detenerse con rojo, pasar con verde). Ignorarlo es una infracción grave."
    }
    
    # Verificación simple de coherencia (si es demasiado corta o no contiene palabras clave)
    if len(respuesta.split()) < 10 or not any(k in respuesta.lower() for k in ['multa', 'accidente', 'conductor']):
        return respaldo_manual.get(pred_label, "Descripción no disponible.")
        
    return respuesta


# =============================================================
# GENERAR Y MOSTRAR DESCRIPCIÓN
# =============================================================
print("\n===================================")
print(f" DESCRIPCIÓN GENERADA PARA: {pred_label.upper()}")
print("===================================\n")

descripcion = generar_descripcion_gpt2(descripcion_base)
print(descripcion)
```
